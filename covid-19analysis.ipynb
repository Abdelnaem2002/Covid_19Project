## Exploratory Data Analysis

```python
df=(df_corona[df_corona['Country'] == 15], 'stemming_text', 20)
df
```

#### **There is a strong relationship between the total deaths and the total cases. The higher the injuries, the higher the deaths

```python
plot.figure(figsize=(8, 10))
sns.lineplot(data = df_corona, y="total_deaths", x="total_cases")
```

```python
plot.figure(figsize=(8, 10))

df_corona['Country'].value_counts().sort_values(ascending=True).head(20).plot(kind='barh',color='blue')
```

#### Question 1 : What country has the big number of injuries and deaths in the world?

```python
top=df_corona.sort_values(by='total_cases').max().head(5)
top
```

####  Question 2 : In what date did Egypt achieve the highest death rate?

```python
df_corona[df_corona.Country=='Egypt'].max()
```

#### Question 3 : What are the top five countries in terms of Deaths?

```python
top_five=df_corona.pivot_table(index='Country',values='total_deaths').sort_values(by='total_deaths',ascending=False)
top_five.head(5)
```

```python

top_five.head(5).plot(kind='line', title='Top Countries with high deaths',color='green')
```

####    Question 4 : On what date was the number of injuries in Egypt exceeded 400 ?

```python
x=df_corona.query(' Country == "Egypt" and total_cases > 400')
x
```
